```python
y=df['Sales']
X=df.drop(columns='Sales')
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,
                                               train_size=0.3,
                                               random_state=667)
```

```python
model=LinearRegression(fit_intercept=True)
model.fit(X_train,y_train)
```

```python
# Beta regressor
model.coef_
```

```python
# Intercept
model.intercept_
```

```python
y_pred=model.predict(X_test)
y_pred_train=model.predict(X_train)
```

```python
plt.figure(figsize = (15, 10))
x = np.linspace(0,y_pred.shape[0], y_pred.shape[0])

#l = y_test_pred.shape[0]
#x = np.linspace(0,l,l)

plt.plot(x,y_pred)
plt.plot(x, y_test)
plt.title('')
plt.xlabel('Rows')
plt.ylabel('Target')
plt.legend(labels= ['y_pred', 'y_test'])
```

```python
res_df = pd.DataFrame(data=list(zip(y_pred, y_test)),columns=['predicted', 'real'])
res_df
```

```python
res_df['error'] = res_df['real'] - res_df['predicted']
res_df
```

```python
res_df['error'].mean()
```

```python

r2score=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=mean_squared_error(y_test,y_pred, squared=False)
    
print('R2 score:',r2score)
print('MAE:' ,mae)
print('MSE:',mse)
print('RMSE:',rmse)
```
