```python
dizio = {"Valle d'Aosta / Vallée d'Aoste": "Valle d'Aosta",
         "Trentino Alto Adige / Südtirol": "Trentino Alto Adige",
         "Provincia Autonoma Bolzano / Bozen": "Bolzano"
         }

df=df.replace({"Territorio": dizio})
```

```python
list_terr = df["Territorio"].unique().tolist()
```

```python
list_terr
```

```python
for i in list_terr:
    df[df["Territorio"] == i].to_csv(f"csvexport/{i}.csv")
    df[df["Territorio"] == i].to_excel(f"xlsxexport/{i}.xlsx")
```
