```python
## split data into several csv by territorio
import pandas as pd

df = pd.read_csv("energia.csv")
diction = {"Valle d'Aosta / Vallée d'Aoste":"Valle_Aosta",
           "Trentino Alto Adige / Südtirol":"Trentino Alto Adige",
           "Provincia Autonoma Bolzano / Bozen":"Bolzano"
            }
df=df.replace({"Territorio": diction})
terri = df["Territorio"].unique().tolist()

for i in terri:
    #print(i)
    df[df["Territorio"]==i].to_csv(f'file_csv/{i}.csv',sep=",",index=False)

print('done!')
```

## Splitting Excel

```python
## split data into several excel by territorio
import pandas as pd

df = pd.read_csv("energia.csv")
diction = {"Valle d'Aosta / Vallée d'Aoste":"Valle_Aosta",
           "Trentino Alto Adige / Südtirol":"Trentino Alto Adige",
           "Provincia Autonoma Bolzano / Bozen":"Bolzano"
            }
df=df.replace({"Territorio": diction})
terri = df["Territorio"].unique().tolist()
for i in terri:
    df[df["Territorio"]==i].to_excel(f'file_excel/{i}.xlsx',index=False)

print('done!')
```

## Merging Excel

```python
import pandas as pd
import glob
import os

# setting the path for joining multiple files
files = os.path.join("file_excel/", "*.xlsx")

# list of merged files returned
files = glob.glob(files)

# joining files with concat and read_csv
df = pd.concat(map(pd.read_excel, files), ignore_index=True,)
df.to_csv("totale1.csv",sep=";",index=False)
```

## Merging CSV

```python
import pandas as pd
import glob
import os

# setting the path for joining multiple files
files = os.path.join("file_csv/", "*.csv")

# list of merged files returned
files = glob.glob(files)

li = []
for filename in files:
    df = pd.read_csv(filename, sep=';', index_col=None, header=0)
    li.append(df)
df = pd.concat(li, axis=0, ignore_index=True)
#df
df.to_csv("totale2.csv",sep=";",index=False)
```

## Sqlite

```python
import pandas as pd
import sqlite3

def populate_sqlite_from_csv(file, database_file,tablename):
    # Leggi il foglio di calcolo Excel in un DataFrame
    df = pd.read_csv(file)
    # Connette al database SQLite
    conn = sqlite3.connect(database_file)
    # Scrivi il DataFrame nel database SQLite
    df.to_sql(tablename, conn, if_exists='replace', index=False)
    # Chiudi la connessione al database
    conn.close()


def populate_sqlite_from_df(df, database_file,tablename):    
    # Connette al database SQLite
    conn = sqlite3.connect(database_file)
    # Scrivi il DataFrame nel database SQLite
    df.to_sql(tablename, conn, if_exists='replace', index=False)
    # Chiudi la connessione al database
    conn.close()

populate_sqlite_from_csv("energia.csv","test.db","energia")
```
